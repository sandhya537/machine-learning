```python
#--------------------    predicting the value  ---------------------------------------


import pandas as pd
from sklearn.linear_model import LinearRegression
df=pd.read_csv('homeprices.csv')

new=df.drop('price',axis=1)

reg=LinearRegression()
reg.fit(new.values,df.price.values)

print(reg.predict([[3300]]))
```
