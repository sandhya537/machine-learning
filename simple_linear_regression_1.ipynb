```python
#----------------------finding coefficent -----------------------------------------------

import pandas as pd
from sklearn.linear_model import LinearRegression
df=pd.read_csv('homeprices.csv')

new=df.drop('price',axis='columns')
price=df.price.values

reg=LinearRegression()
reg.fit(new.values,price)

print(reg.coef_)
```

```python
#--------------------  finding intercept ----------------------------------------------------------

import pandas as pd
from sklearn.linear_model import LinearRegression
df=pd.read_csv('homeprices.csv')

new=df.drop('price',axis=1)
price=df.price.values

reg=LinearRegression()
reg.fit(new.values,price)

print(reg.intercept_)
```

```python
#-----------------------------  regression line---------------------------------------

import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df=pd.read_csv('homeprices.csv')

new=df.drop('price',axis='columns')
price=df.price.values

reg=LinearRegression()
reg.fit(new.values,price)

plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
plt.show()
```

```python
#------------------     practice 2   -------------------------------------------------

import pandas as pd
from sklearn.linear_model import LinearRegression

df=pd.read_csv('homeprices.csv')

new=df.drop('price',axis=1)
price=df.price.values

reg=LinearRegression()
reg.fit(new.values,price)

area_df=pd.read_csv('areas.csv')
pri=reg.predict(area_df.values)

area_df['price']=pri
area_df.to_csv('output.csv')

print('please check in current directory output.csv')

```
