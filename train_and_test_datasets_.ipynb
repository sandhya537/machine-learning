```python
#------------------------------  using train ,test 2 datasets -------------------------------------

import numpy as np
from sklearn.model_selection import train_test_split

X=np.arange(20).reshape(10,2)
print(X)

Y=np.arange(10)
print(Y)
print()
X_train,X_test,Y_train,Y_test=train_test_split(X,Y)

print(X_train)
print(X_test)
print()
print(Y_train)
print(Y_test)
```

```python
#------------------------------  using 3 datasets for test,train   -------------------------------------
#-------------------practice 1-----------------------

import numpy as np
from sklearn.model_selection import train_test_split

X=np.arange(10)
print(X)

Y=np.arange(10)
print(Y)

Z=np.arange(10)
print(Z)

print()
X_train,X_test,Y_train,Y_test,Z_train,Z_test=train_test_split(X,Y,Z)
print(X_train)
print(X_test)
print()
print(Y_train)
print(Y_test)
print()
print(Z_train)
print(Z_test)


```

```python
#------------------------------  using random_state=0  -------------------------------------

import numpy as np
from sklearn.model_selection import train_test_split

X=np.arange(20).reshape(2,10).T
print(X)

Y=np.arange(10)
print(Y)

print()
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=0)

print(X_train)
print(X_test)
print()
print(Y_train)
print(Y_test)
```

```python
#------------------------------  using 2 data sets -------------------------------------
#-------------------practice 2 reshape-----------------------

import numpy as np
from sklearn.model_selection import train_test_split

X=np.arange(10).reshape(10,1)
print(X)

Y=np.arange(10)
print(Y)
print()
X_train,X_test,Y_train,Y_test=train_test_split(X,Y)

print(X_train)
print(X_test)
print()
print(Y_train)
print(Y_test)
```
